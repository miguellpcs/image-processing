Another powerful operation is the **Einsum**. Based on the Einstein notation of sumation, this function can be used to model a plenty of mathematical operations just by indexing. This function is well-optimized and outperforms any python implementation of the operations, but its very important to learn how to use it, since its sintax can be very tricky. 

Examples

```python
a = np.arange(0,11)
b = np.arange(23,34)
```

Sum

```python
np.einsum('i->', a)
```

Element-wise multiplication


```python
np.einsum('i,i->i', a,b)
```

Dot product

```python
np.einsum('i,i->',a,b)
```

### Chapter 3 - Intensity Transformations and Spatial Filtering

**Image Negatives:**
Is produced by subtracting each pixel from the maximum intensity value, e.g., for a 8-bit image, the max value (L) is 255. 
The transformation function used is **s = T(r) = L - 1 - r**

```python
def neg_transform(img):
    L = np.max(img)
    img_neg = L - 1 - img
    return img_neg

img_neg = neg_transform(img3)

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(2, 2,  1)
plt.title('Original',y=-0.5,color='r')
plt.imshow(img3)


plt.subplot(2, 2,  2)
plt.title('Negative',y=-0.5,color='r')
plt.imshow(img_neg)

plt.show()
```

**Log Transformations**: 

s = T(r) = c*log(1 + r) where c is a constant.

```python
img_log = 1 + img
img_log.min()
```

The array we've created from this image file has a data type of uint8, with a maximum value of 255. When adding positive numbers to the array, some elements will roll over this value. In particular, any pixel's colour channel that had the value 255 will be 0 when we add 1 to it.

```python
cs = [0.1, 0.2, 1]

img_float = img.astype(float)
img_float = 1 + img_float

plt.figure(figsize=(6, 7), dpi=80)
plt.subplot(2, 2,  1)
plt.title('Original',y=-0.5,color='r')
plt.imshow(img)

img_logs = [img_float]

for c in (cs):
  img_logs.append(c*np.log(img_float))
  
for i,im in enumerate(img_logs):
  plt.subplot(2, 2, 1+i)
  plt.title(f'Image Log, c = {c}',y=-0.5,color='r')
  plt.imshow(im)




plt.show()
```
